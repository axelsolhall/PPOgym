```python
import os
import subprocess

# Ensure directories exist
os.makedirs("frames", exist_ok=True)
os.makedirs("gifs", exist_ok=True)

# Process videos into smooth GIFs
for file in os.listdir("videos"):
    if file.endswith(".mp4"):
        filename = file.split(".")[0]
        
        # Step 1: Extract frames from the video
        extract_command = f"ffmpeg -y -i videos/{file} -vf fps=15,scale=320:-1 frames/{filename}_%04d.png"
        subprocess.run(extract_command, shell=True)
        
        # Step 2: Create GIF with explicit frame delay
        delay_ms = 6.67  # 15 FPS = ~6.67ms delay per frame
        gif_command = (
            f"magick -delay {delay_ms} -loop 0 frames/{filename}_*.png gifs/{filename}.gif"
        )
        subprocess.run(gif_command, shell=True)
        
        # Step 3: Clean up extracted frames
        for frame in os.listdir("frames"):
            if frame.startswith(filename):
                os.remove(os.path.join("frames", frame))
```
